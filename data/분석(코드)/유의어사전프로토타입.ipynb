```python
import pandas as pd 
import numpy as np 

df = pd.read_csv('유의어사전라벨링.csv',index_col=0)
```

```python
from collections import defaultdict 
dic = {}
for i in df.index:
    name = i.strip()
    dic[name] = []
    
df_label = pd.DataFrame(dic).transpose()

for i in range(5):
    df_label[str(i)] = 0
```

```python
for name in df.index:
    label = df.loc[name,'Label'] 
    if label != '-':
        name = name.strip()
        if type(label) != float :
            label = label.split(',')
            for num in label:
                df_label.loc[name,num] = 1
        else:
            df_label.loc[name,label] = 1
```
